### Heroes Of Pymoli 

Three trends based on observable data:

* The majority of customers identify as Male: 84% of 576 screen names. 

* The peak age demographic is college-aged: between 20-24 (44.8%).

* Female customers are willing to spend more than their Male counterparts, on average. And college-aged customers spend more than other age groups.

* Note: One noteable drawback to the provided data is that we cannot know if players often use more than one screen name, in which case their demographic data would be duplicated.
-----

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
player_names = purchase_data["SN"].unique()
total_players = len(player_names)

total_players_df = pd.DataFrame({'Total Players':[total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())

avg_price = purchase_data["Price"].mean()
avg_price_string = '${:,.2f}'.format(avg_price)

purchase_count = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()
total_revenue_string = '${:,.2f}'.format(total_revenue)

summary = [{"Number of Unique Items": unique_items, 
            "Average Price": avg_price_string, 
            "Number of Purchases": purchase_count, 
            "Total Revenue": total_revenue_string}]
summary_df = pd.DataFrame(summary)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
small_df = purchase_data[['SN', 'Gender', 'Age']]
small_df_nodups = small_df.drop_duplicates()
gender = small_df_nodups["Gender"].value_counts()

gender_df = pd.DataFrame([{"Gender":"Male", "Total Count":gender["Male"],
                           "Percentage of Players":((gender["Male"]/total_players)*100)},
                          {"Gender":"Female", "Total Count":gender["Female"], 
                           "Percentage of Players":((gender["Female"]/total_players)*100)},
                          {"Gender":"Other / Non-Disclosed", "Total Count":gender["Other / Non-Disclosed"],
                           "Percentage of Players":((gender["Other / Non-Disclosed"]/total_players)*100)}])
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bygender = purchase_data.groupby(['Gender'])

purchasecount = bygender["Purchase ID"].count()
avgpurchaseprice = bygender["Price"].mean()
totalpurchaseprice = bygender["Price"].sum()
gender = small_df_nodups["Gender"].value_counts()
avgpurchaseperperson = totalpurchaseprice/gender

gender_analysis_df = pd.DataFrame({"Purchase Count":purchasecount, 
                                   "Average Purchase Price":avgpurchaseprice.map("${:.2f}".format),
                                   "Total Purchase Value":totalpurchaseprice.map("${:.2f}".format),
                                   "Avg Total Purchase Per Person":avgpurchaseperperson.map("${:.2f}".format)})
gender_analysis_df
                                   


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

dfnodups = small_df_nodups.copy()
dfnodups["Age Range"] = pd.cut(dfnodups["Age"], bins, labels=group_names)

groupedbyage = dfnodups.groupby(["Age Range"])

totalcountbyage = groupedbyage["Age"].count()
percentbyage = ((totalcountbyage/total_players) *100).map("{:.2f}".format)

age_analysis_df = pd.DataFrame({"Total Count":totalcountbyage,
                                "Percentage of Players":percentbyage})
age_analysis_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
dfcopy = purchase_data.copy()
bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

dfcopy["Age Range"] = pd.cut(dfcopy["Age"], bins, labels=group_names)

copybyage = dfcopy.groupby(["Age Range"])

purchasecountbyage = copybyage["SN"].count()
avgpurchasepricebyage = copybyage["Price"].mean()
totalpurchasevaluebyage = copybyage["Price"].sum()
avgtotalpurchaseperperson = totalpurchasevaluebyage/groupedbyage["Age"].count()

purchase_analysis_by_age = pd.DataFrame({"Purchase Count":copybyage["SN"].count(), 
                                         "Average Purchase Price":copybyage["Price"].mean().map("${:.2f}".format),
                                         "Total Purchase Value":copybyage["Price"].sum().map("${:.2f}".format),
                                         "Avg Total Purchase Per Person":avgtotalpurchaseperperson.map("${:.2f}".format)
                                        })
purchase_analysis_by_age.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
copybySN = dfcopy.groupby(["SN"])
purchasecountbySN = copybySN["Purchase ID"].count()
avgpurchasepricebySN = copybySN["Price"].mean()
totalpurchasevaluebySN = copybySN["Price"].sum()

purchase_analysis_by_SN = pd.DataFrame({"Purchase Count":copybySN["Purchase ID"].count(), 
                                         "Average Purchase Price":copybySN["Price"].mean().map("${:.2f}".format),
                                         "Total Purchase Value":copybySN["Price"].sum().map("${:.2f}".format),
                                        })
purchase_analysis_by_SN.sort_values(["Total Purchase Value"], ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
populardf = purchase_data[["Item ID", "Item Name", "Price"]]
populargrouped = populardf.groupby(["Item ID", "Item Name"])

poppurchasecount = populargrouped["Price"].count()

poptotalpurchasevalue = populargrouped["Price"].sum()

itemprice = populargrouped["Price"].mean()

popular_df = pd.DataFrame({"Purchase Count":poppurchasecount,
                           "Item Price":itemprice.map("${:.2f}".format),
                           "Total Purchase Value":poptotalpurchasevalue.map("${:.2f}".format)})

popular_df.sort_values(["Purchase Count"], ascending=False).head()
```
